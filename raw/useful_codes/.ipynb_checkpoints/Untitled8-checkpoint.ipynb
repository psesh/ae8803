```python
import gpflow
import gpflux
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
n = 80
sigma = 0.05
X = np.random.rand(n)*2 - 1.
Y = np.sign(X) + np.random.randn(n) * sigma
X = X.reshape(n,1)
Y = Y.reshape(n,1)
X_test = np.linspace(-2, 2, 150).reshape(150,1)
```

```python
fig = plt.figure(figsize=(5,2))
plt.plot(X, Y, '.')
plt.show()
```

```python
kernel = gpflow.kernels.SquaredExponential()
mu = gpflow.mean_functions.Zero()

inducing = gpflow.inducing_variables.InducingPoints(X)

# First layer of the DGP!
gp_layer = gpflux.layers.GPLayer(kernel, \
                num_data=n,
                inducing_variable=inducing,
                mean_function=mu,
                num_latent_gps=1)

# Likelihood aka noise model!
noise_model = gpflow.likelihoods.Gaussian(variance=sigma**2)
likelihood_layer = gpflux.layers.LikelihoodLayer(noise_model)
```

```python
single = gpflux.models.DeepGP([gp_layer, gp_layer], \
                              likelihood_layer)
```

```python
model = single.as_training_model()
model.compile(tf.optimizers.Adam(0.01))
```

```python
hist = model.fit({"inputs":X, "targets": Y}, \
                epochs=int(1e3), verbose=0)
plt.plot(hist.history["loss"])
```

```python
model_out = single.as_prediction_model()
out = model_out(X_test.reshape(150,1))
```

```python
mu_ = out.f_mean.numpy().squeeze()
var_ = out.f_var.numpy().squeeze()
std_ = np.sqrt(var_)
```

```python
X_test = X_test.squeeze()
```

```python
fig = plt.figure(figsize=(6,2))
plt.plot(X_test, mu_, 'r.')
plt.plot(X, Y, '.', alpha=0.4)
plt.fill_between(X_test, mu_ + std_, mu_ - std_, alpha=0.3)
```
