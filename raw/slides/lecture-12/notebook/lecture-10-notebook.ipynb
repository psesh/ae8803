```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from scipy.linalg import cholesky, solve_triangular
import seaborn as sns
sns.set(font_scale=1.0)
sns.set_style("white")
sns.set_style("ticks")
palette = sns.color_palette('deep')
plt.style.use('dark_background')
```

```python
t = np.random.rand(40,1)* 2 * np.pi
r = np.random.rand(40,1)*0.2 + 2
u = r * np.cos(t)
v = r * np.sin(t)

xy = np.vstack([np.random.rand(20,2)*2 - 1])
xy2 = np.hstack([u, v])

def mapup(xy):
    z1 = xy[:,0]**2
    z2 = np.sqrt(2) * xy[:,0] * xy[:,1]
    z3 = xy[:,1]**2
    return z1, z2, z3

z1, z2, z3 = mapup(xy2)
w1, w2, w3 = mapup(xy)

```

```python
fig, ax = plt.subplots(2, figsize=(10,3))
fig.patch.set_facecolor('#6C757D')
ax = fig.add_subplot(1, 2, 1)
ax.set_fc('#6C757D')
plt.subplot(121)
plt.plot(xy2[:,0], xy2[:,1], 'ro')
plt.plot(xy[:,0], xy[:,1], 'gs')
plt.title(r'Two dimensional space, ($\mathbf{x}$, $\mathbf{v}$)')
plt.xlabel(r'$\mathbf{x}$')
plt.ylabel(r'$\mathbf{v}$')
fig.patch.set_facecolor('#6C757D')

ax = fig.add_subplot(1, 2, 2, projection='3d')
plt.rcParams['axes.facecolor']='#6C757D'
ax.set_facecolor('#6C757D')
ax.plot(z1, z2, z3, 'ro')
ax.plot(w1, w2, w3, 'gs')
ax.view_init(10, 95)
plt.title(r'Three dimensional space, ($\phi_1$, $\phi_2$, $\phi_3$)')
plt.savefig('spaces.png', dpi=150, bbox_inches='tight', facecolor="#6C757D")
plt.show()
```
